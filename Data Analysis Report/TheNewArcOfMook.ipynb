```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data for Refund High School Metrics
data = {
    'Metric': ['Page Views', 'Average Time Spent (min)', 'Bounce Rate (%)'],
    'Value': [8000, 4, 35]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# Bar Chart for Refund High School Metrics
plt.figure(figsize=(8, 5))
plt.bar(df['Metric'], df['Value'], color='lightcoral')
plt.title('Refund High School Performance Metrics')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
```

```python
# Data for Bounce Rate Pie Chart
bounce_data = {
    'Bounce Rate': [35, 65],  # 35% Bounce Rate, 65% Non-Bounce
    'Labels': ['Bounce', 'Non-Bounce']
}

bounce_df = pd.DataFrame(bounce_data)
```

```python
bounce_df
```

```python
# Pie Chart for Bounce Rate Distribution
plt.figure(figsize=(6, 6))
plt.pie(bounce_df['Bounce Rate'], labels=bounce_df['Labels'], autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
plt.title('Bounce Rate Distribution for Refund High School')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for Refund High School Metrics
data = {
    'Metric': ['Page Views', 'Average Time Spent (min)', 'Bounce Rate (%)'],
    'Value': [8000, 4, 35]
}

df = pd.DataFrame(data)

# Bar Chart for Refund High School Metrics
plt.figure(figsize=(8, 5))
plt.bar(df['Metric'], df['Value'], color='lightcoral')
plt.title('Refund High School Performance Metrics')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()

# Data for Bounce Rate Pie Chart
bounce_data = {
    'Bounce Rate': [35, 65],  # 35% Bounce Rate, 65% Non-Bounce
    'Labels': ['Bounce', 'Non-Bounce']
}

bounce_df = pd.DataFrame(bounce_data)

# Pie Chart for Bounce Rate Distribution
plt.figure(figsize=(6, 6))
plt.pie(bounce_df['Bounce Rate'], labels=bounce_df['Labels'], autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
plt.title('Bounce Rate Distribution for Refund High School')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()
```
