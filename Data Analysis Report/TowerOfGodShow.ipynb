```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Metric': ['Page Views', 'Average Time Spent (min)', 'Bounce Rate (%)'],
    'Value': [10000, 3, 45]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(df['Metric'], df['Value'], color='skyblue')
plt.title('Page Views vs Average Time Spent')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
```

```python
# Pie Chart Data
bounce_data = {
    'Bounce Rate': [45, 55],
    'Labels': ['Bounce', 'Non-Bounce']
}

bounce_df = pd.DataFrame(bounce_data)
```

```python
bounce_df
```

```python
bounce_df = pd.DataFrame(bounce_data)

# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(bounce_df['Bounce Rate'], labels=bounce_df['Labels'], autopct='%1.1f%%', startangle=90)
plt.title('Bounce Rate Distribution')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for Bar Chart
data = {
    'Metric': ['Page Views', 'Average Time Spent (min)', 'Bounce Rate (%)'],
    'Value': [10000, 3, 45]
}

df = pd.DataFrame(data)

# Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(df['Metric'], df['Value'], color='skyblue')
plt.title('Page Views vs Average Time Spent')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()

# Sample data for Pie Chart
bounce_data = {
    'Bounce Rate': [45, 55],
    'Labels': ['Bounce', 'Non-Bounce']
}

bounce_df = pd.DataFrame(bounce_data)

# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(bounce_df['Bounce Rate'], labels=bounce_df['Labels'], autopct='%1.1f%%', startangle=90)
plt.title('Bounce Rate Distribution')
plt.axis('equal')
plt.show()
```
