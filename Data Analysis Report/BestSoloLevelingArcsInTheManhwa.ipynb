```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data for Solo Leveling Performance Metrics
data = {
    'Metric': ['Page Views', 'Average Time Spent (min)', 'Bounce Rate (%)'],
    'Value': [12000, 6, 25]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# Bar Chart for Solo Leveling Performance Metrics
plt.figure(figsize=(8, 5))
plt.bar(df['Metric'], df['Value'], color='steelblue')
plt.title('Solo Leveling Performance Metrics')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
```

```python
# Data for User Segmentation (by Age Group)
age_group_data = {
    'Age Group': ['18-24', '25-34', '35-44', '45+'],
    'Percentage': [40, 35, 15, 10]
}

age_group_df = pd.DataFrame(age_group_data)
```

```python
age_group_df
```

```python
# Pie Chart for User Segmentation by Age Group
plt.figure(figsize=(6, 6))
plt.pie(age_group_df['Percentage'], labels=age_group_df['Age Group'], autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'orange', 'lightblue', 'pink'])
plt.title('User Segmentation by Age Group for Solo Leveling')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for Solo Leveling Performance Metrics
data = {
    'Metric': ['Page Views', 'Average Time Spent (min)', 'Bounce Rate (%)'],
    'Value': [12000, 6, 25]
}

df = pd.DataFrame(data)

# Bar Chart for Solo Leveling Performance Metrics
plt.figure(figsize=(8, 5))
plt.bar(df['Metric'], df['Value'], color='steelblue')
plt.title('Solo Leveling Performance Metrics')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()

# Data for User Segmentation (by Age Group)
age_group_data = {
    'Age Group': ['18-24', '25-34', '35-44', '45+'],
    'Percentage': [40, 35, 15, 10]
}

age_group_df = pd.DataFrame(age_group_data)

# Pie Chart for User Segmentation by Age Group
plt.figure(figsize=(6, 6))
plt.pie(age_group_df['Percentage'], labels=age_group_df['Age Group'], autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'orange', 'lightblue', 'pink'])
plt.title('User Segmentation by Age Group for Solo Leveling')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()
```
